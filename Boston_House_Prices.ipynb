# Challenge - Boston House Prices

## 1. Data Preparation

```python
# Import and load the dataset
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
X,y = load_boston(return_X_y=True)
```

```python
# Have a look at the data
X.shape,y.shape
```

```python
# Let's split and rescale
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test,y_train, y_test= train_test_split(X, y, test_size=0.3,random_state=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape
```

## 2. Modelling

Now we want to build a Neural Network to perform this regression task. We will build a Neural Networks with 5 hidden layers of 100 units each.

```python
# Create the Sequential object
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Dense(units=100,input_dim=13,activation='relu'),
                                    tf.keras.layers.Dense(units=100,activation='relu'),
                                    tf.keras.layers.Dense(units=100,activation='relu'),
                                    tf.keras.layers.Dense(units=100,activation='relu'),
                                    tf.keras.layers.Dense(units=100,activation='relu')

])
```

Now let's compile the model and then fit it

```python
# Compile the model with mean squared error (for regression)

# Now fit the model on 500 epoches with a batch size of 64
# You can add the test/validation set into the fit: it will give insights on this dataset too
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["mean_squared_error"])
```

```python
model.fit(X_train,y_train, validation_data=(X_test, y_test), epochs=500,batch_size=64, )
```

Finally we can compute the results of our model

```python
# Evaluation
model.predict(X_test)
loss,mean_squared_error= model.evaluate(X_test,y_test,verbose=0)
```

```python
print('loss is:', loss)
print('mean squared error is:', mean_squared_error)
```

## 3. Benchmarking

We can compare that result to a good old linear regression

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mse
```
